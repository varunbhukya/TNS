MDP is the mathematical foundation of RL.

It contains:

States (S)
Actions (A)
Rewards (R)
Transition Probability (P)
Discount Factor (γ)

Formula

V(s)=R+γV(next)

```python
states = ["A","B","Goal"]

rewards = {
    "A":0,
    "B":1,
    "Goal":10
}

gamma = 0.9

value_goal = rewards["Goal"]

value_B = rewards["B"] + gamma*value_goal

value_A = rewards["A"] + gamma*value_B

print("Value of Goal =",value_goal)
print("Value of B =",value_B)
print("Value of A =",value_A)
```
