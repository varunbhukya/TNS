```python
student_names=[]
subjects=set()
marks_dict={}

name=input("enter student name:")
age=int(input("enter age:"))
student_information=(name, age)

number_of_subjects=int(input("enter number of subjects:"))
student_names.append(name)

total_marks = 0
for i in range(number_of_subjects):
  subject=input(f"enter {i+1} subject name:") # Corrected f-string for i+1
  marks=int(input("enter marks:"))
  subjects.add(subject)
  marks_dict[subject]=marks # Correctly store subject-wise marks
  total_marks += marks

# Calculate average marks
if number_of_subjects > 0:
  average_marks = total_marks / number_of_subjects
else:
  average_marks = 0

# Print Report Card
print("\n--- Student Report Card ---")
print(f"Name: {student_information[0]}")
print(f"Age: {student_information[1]}")
print("Subjects and Marks:")
for sub, mks in marks_dict.items():
  print(f"  {sub}: {mks}")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks:.2f}")
```
