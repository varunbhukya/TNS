```python
import matplotlib.pyplot as plt

students = ['A','B','C','D','E','F','G','H','I','J']
marks = [78,85,92,67,88,95,72,81,90,76]
plt.figure(figsize=(8,6))
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
max_marks = max(marks)
top_student = students[marks.index(max_marks)]
print("Highest Scorer:", top_student, "-", max_marks)
sorted_data = sorted(zip(students, marks), key=lambda x: x[1], reverse=True)
students_desc, marks_desc = zip(*sorted_data)

plt.figure(figsize=(8,6))
plt.bar(students_desc, marks_desc)
plt.title("Student Marks (Descending Order)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
