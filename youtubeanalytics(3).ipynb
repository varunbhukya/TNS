```python
import kagglehub
import pandas as pd
import os
import matplotlib.pyplot as plt

path = kagglehub.dataset_download("nelgiriyewithana/global-youtube-statistics-2023")

print("Path to dataset files:", path)

print("Contents of the dataset directory:", os.listdir(path))

csv_filename = "Global YouTube Statistics.csv"
full_csv_path = os.path.join(path, csv_filename)

df = pd.read_csv(full_csv_path, encoding='latin1')
country_creator_counts = df['Country'].value_counts().head(20)
Country_for_second_plot = country_creator_counts.index.tolist()
Youtuber_counts_for_plot = country_creator_counts.values.tolist()

plt.figure(figsize=(10,6))
plt.bar(Country_for_second_plot, Youtuber_counts_for_plot)
plt.title("Top 20 Countries by Number of Youtubers")
plt.xlabel("Country")
plt.ylabel("Number of Youtubers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

top_20_channels = df.sort_values(by='subscribers', ascending=False).head(20)

Channel_for_third_plot = top_20_channels['Youtuber'].tolist()
Subscribers_counts_for_plot = top_20_channels['subscribers'].tolist()

plt.figure(figsize=(10,6))
plt.bar(Channel_for_third_plot, Subscribers_counts_for_plot)
plt.title("Top 20 channels by Subscribers")
plt.xlabel("Channel")
plt.ylabel("Subscribers")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

top_20_channels = df.sort_values(by='video views', ascending=False).head(20)

Channel_for_third_plot = top_20_channels['Youtuber'].tolist()
Subscribers_counts_for_plot = top_20_channels['video views'].tolist()

plt.figure(figsize=(10,6))
plt.bar(Channel_for_third_plot, Subscribers_counts_for_plot)
plt.title("Top 20 channels with highest views")
plt.xlabel("Channel")
plt.ylabel("views")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
path = kagglehub.dataset_download("nelgiriyewithana/global-youtube-statistics-2023")

print("Path to dataset files:", path)

print("Contents of the dataset directory:", os.listdir(path))

csv_filename = "Global YouTube Statistics.csv"
full_csv_path = os.path.join(path, csv_filename)

df = pd.read_csv(full_csv_path, encoding='latin1')

df.nlargest(10, 'subscribers').plot(
    x='Youtuber', y='subscribers', kind='bar',
    figsize=(8,4), title='Top Channels by Subscribers'
)
plt.xticks(rotation=90, ha='right')
plt.show()

df.nlargest(10, 'video views').plot(
    x='Youtuber', y='video views', kind='bar',
    figsize=(8,4), title='Top Channels by Video Views'
)
plt.xticks(rotation=90, ha='right')
plt.show()

df['Country'].value_counts().head(10).plot(
    kind='bar', figsize=(8,4),
    title='Top Countries by Number of Creators'
)
plt.xticks(rotation=90, ha='right')
plt.show()
```
