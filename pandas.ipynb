```python
import pandas as pd
import numpy as np
s=pd.Series([1,2,np.nan,4,5,6])

print(s)
all=s.describe()
print(all)
mean=s.mean()
print(mean)
loc=s.iloc[-2]
print(loc)
locc=s.loc[5]
print(locc)
filter = s[s >= 2]
print(filter)
filterr=s[s<=2]
print(filterr)
drop=s.drop(2)
print(drop)
null=s.isnull()
print(null)
notnull=s.notnull()
print(notnull)
sum=s.sum()
print(sum)
cum_sum=s.cumsum()
print(cum_sum)
aggregate = s.agg(['sum', 'mean'])
print(aggregate)
```
