Anomaly Detection identifies unusual data points that differ significantly from the majority of the data.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Create normal data
np.random.seed(42)
normal_data = np.random.randn(200, 2)

# Create anomaly points
anomalies = np.random.uniform(low=-6, high=6, size=(20,2))

# Combine data
X = np.vstack((normal_data, anomalies))

# Train Isolation Forest
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(X)

# Predict
predictions = model.predict(X)

# Plot
plt.figure(figsize=(8,6))

plt.scatter(
    X[predictions==1,0],
    X[predictions==1,1],
    color='blue',
    label='Normal'
)

plt.scatter(
    X[predictions==-1,0],
    X[predictions==-1,1],
    color='red',
    label='Anomaly'
)

plt.title("Anomaly Detection using Isolation Forest")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
